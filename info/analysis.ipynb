```python
import matplotlib.pyplot as plt
import pandas as pd
epochs = [i for i in range(1, 101)]
data = pd.read_csv('./slice5.txt', names=["accuracy_train", "loss_train", "accuracy_val", "loss_val"])
accuracy_train = data["accuracy_train"].values
accuracy_val = data["accuracy_val"].values
loss_train = data["loss_train"]
loss_val = data["loss_val"]

fig, ax = plt.subplots(1,2,figsize=(12,4))
# V  th
ax[0].plot(epochs, accuracy_train, linestyle='-', color='b', label = "Accuracy Train")
ax[0].plot(epochs, accuracy_val, linestyle='-', color='orange', label = "Accuracy Validation")

# t tiu  v nhn cho trc x, y
ax[0].set_xlabel('Epochs', fontsize = 12)
ax[0].set_ylabel('Accuracy', fontsize = 12)
ax[0].set_title('Accuracy on Fold5', fontsize = 15)
ax[0].legend(loc = 'lower right')

# V  th
ax[1].plot(epochs, loss_train, linestyle='-', color='b', label = "Loss Train")
ax[1].plot(epochs, loss_val, linestyle='-', color='orange', label = "Loss Validation")

# t tiu  v nhn cho trc x, y
ax[1].set_xlabel('Epochs', fontsize = 12)
ax[1].set_ylabel('Loss', fontsize = 12)
ax[1].set_title('Loss on Fold5', fontsize = 15)
ax[1].legend(loc = 'upper right')
# plt.subplots_adjust(wspace=0.2)
plt.show()
```
